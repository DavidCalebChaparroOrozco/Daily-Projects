```python
# Importing necessary libraries
import pandas as pd
from string import ascii_uppercase as alphabet
import pickle
```

# Reading all tables from the Wikipedia page for the 2024 Copa América

```python
# Reading all tables from the Wikipedia page for the 2024 Copa América
all_tables = pd.read_html("https://en.wikipedia.org/wiki/2024_Copa_Am%C3%A9rica")
```

## Iterate over letters A to D and the corresponding table indices


```python
for letter, i in zip(alphabet,range(15,43, 7)):
    print(letter, i)
```

```python
# Creating a dictionary to store the tables for each group
dict_tables = {}

# Iterate over letters A to D and the corresponding table indices
for letter, i in zip(alphabet,range(15,43, 7)):
    data = all_tables[i]
    # data.rename(columns={"Teamvte": "Team"})
    # Rename the second column to "Team"
    data.rename(columns={data.columns[1]: "Team"}, inplace=True)
    # Remove the "Qualification" column
    data.pop("Qualification")
    # Store the table in the dictionary with the group letter as the key
    dict_tables[f'Group {letter}'] = data
    print(data)
```

```python
# Display the keys of the dictionary
dict_tables.keys()
```

```python
# Display the table for Group A
dict_tables['Group A']
```

```python
# Save the dictionary to a file using pickle
with open('dict_table', 'wb') as file:
    pickle.dump(dict_tables, file)
```
