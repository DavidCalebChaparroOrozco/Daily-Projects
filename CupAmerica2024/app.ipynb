```python
# Import necessaries libraries
from bs4 import BeautifulSoup
import requests
```

```python
web = "https://es.wikipedia.org/wiki/Copa_Am%C3%A9rica_2021"
response = requests.get(web)
print(response.text)
```

```python
content = response.text
soup = BeautifulSoup(content, "lxml")
```

```python
matches = soup.find_all("table", class_="collapsible autocollapse vevent plainlist mw-collapsible mw-made-collapsible mw-collapsed")
```

```python
for match in matches:
    match.find()
```

```python
# Iterar sobre todos los elementos <tr>
for tr in soup.find_all('tr'):
    # Encontrar el equipo local
    home_team_link = tr.find('a', {'title': True})
    if home_team_link:
        home_team = home_team_link['title'].split(' de ')[1]
    else:
        home_team = 'Unknown'
    
    # Encontrar el equipo visitante
    away_team_links = tr.find_all('a', {'title': True})
    if len(away_team_links) > 1:
        away_team = away_team_links[-1]['title'].split(' de ')[1]
    else:
        away_team = 'Unknown'
    
    # Encontrar el marcador
    score_div = tr.find('div', style='white-space:nowrap;')
    if score_div:
        score = score_div.text.strip()
    else:
        score = 'Unknown'
    
    print(f"home = {home_team}")
    print(f"score = {score}")
    print(f"away = {away_team}")
    print()
```
